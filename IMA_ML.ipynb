```python
import pandas as pd
import datetime
import numpy as np
from sklearn.preprocessing import OneHotEncoder

data = pd.read_csv('DATA.csv', header = [2,3])

data.columns = data.columns.droplevel(1)

data = data[:-1]

data.head()
```

```python
data.columns.values
```

```python
df = data[['Дата','Скважина', 'Пласт', 'Куст', 'Пластовое давление', 'Коэффициент продуктивности']].copy()

for index, row in enumerate(df['Дата']):
    temp_str = row.split('.')
    if len(temp_str[0]) == 2:
        date = datetime.datetime.strptime('.'.join(temp_str), '%m.%Y')
    else:
        date = datetime.datetime.strptime('.'.join(temp_str), '%Y.%m')
    df.loc[index:index,('Дата')] = date
    
df[['Пластовое давление','Коэффициент продуктивности']] = df[['Пластовое давление','Коэффициент продуктивности']].replace(",", '.', regex=True)
df['Пластовое давление'] = np.array(df['Пластовое давление']).astype(np.float64)
df['Коэффициент продуктивности'] = np.array(df['Коэффициент продуктивности']).astype(np.float64)

# df['Скважина']
# df['Пласт']
# df['Куст']

df = df.groupby(['Дата', 'Скважина', 'Пласт','Куст']).mean().reset_index()

df.head()
```
